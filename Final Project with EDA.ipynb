```python
# --- 1. Import Dependencies ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, KFold # Added KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
# Removed: from scikeras.wrappers import KerasRegressor
import warnings
import traceback
import time
import itertools # For manual grid generation

# Suppress warnings and TF logs
warnings.filterwarnings('ignore')
tf.get_logger().setLevel('WARN')
tf.autograph.set_verbosity(0)

print("Dependencies imported.")

# --- Configuration ---
FILE_PATH = "full_data_set_merged.csv"
DATE_COLUMN = "DATE"
ECONOMIC_INDICATORS = ["PRICE_PER_GALLON", "UNEMPLOYMENT_RATE", "AVG_MORTGAGE_RATE", "INTEREST_RATES", "INFLATION"]
TARGET_PREFIX = ""
MODEL_FEATURES = ["PRICE_PER_GALLON", "UNEMPLOYMENT_RATE", "INTEREST_RATES", "INFLATION"]
CV_FOLDS = 3 # Number of cross-validation folds for manual NN grid search and RF GridSearchCV
NN_MANUAL_EPOCHS = 50 # Fixed epochs for manual grid search training (can be tuned)
NN_MANUAL_BATCH_SIZE = 32 # Fixed batch size for manual grid search (can be tuned)
FINAL_NN_EPOCHS = 100 # Epochs for training the final best NN model

# --- Function Definitions ---

# --- 1.1 to 1.10: Keep existing functions ---
# load_inspect_convert_dates, visualize_time_series, show_descriptive_stats,
# visualize_distributions, analyze_correlation, visualize_pairplot,
# get_target_region_input, visualize_target_relationships,
# prepare_data_for_model, split_data
# (Functions are the same as in the previous version)
# --- 1.1 Data Loading and Initial Inspection (with Date Conversion) ---
def load_inspect_convert_dates(file_path, date_col):
    """ Loads data, inspects, and converts date column. """
    print("\n--- 1. Data Loading, Inspection, and Date Conversion ---")
    try:
        df = pd.read_csv(file_path)
        print(f"Data loaded successfully from: {file_path}")

        if date_col not in df.columns:
            print(f"Error: Specified date column '{date_col}' not found.")
            potential_dates = df.select_dtypes(include=['object', 'string']).columns
            if len(potential_dates) == 1 and ('date' in potential_dates[0].lower() or 'time' in potential_dates[0].lower()):
                print(f"Attempting to use column '{potential_dates[0]}' as date column.")
                date_col = potential_dates[0]
            else:
                print("Cannot proceed without a valid date column.")
                return None

        print("\nBasic Info (Before Date Conversion):")
        df.info(verbose=False); print("\nFirst 5 rows:"); print(df.head())
        print("\nMissing Values per Column (Top 10):")
        missing_counts = df.isnull().sum()
        print(missing_counts[missing_counts > 0].sort_values(ascending=False).head(10))

        print(f"\nConverting '{date_col}' column to datetime...")
        original_len = len(df)
        df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
        na_count = df[date_col].isna().sum()
        if na_count > 0:
            print(f"Warning: Could not parse {na_count} out of {original_len} dates. Dropping rows with invalid dates.")
            df.dropna(subset=[date_col], inplace=True)

        df = df.sort_values(by=date_col).reset_index(drop=True)
        print(f"Data sorted by '{date_col}'.")
        print("\nBasic Info (After Date Conversion):"); df.info(verbose=False)
        return df
    except FileNotFoundError: print(f"Error: File not found at {file_path}"); return None
    except Exception as e: print(f"An error occurred: {e}"); traceback.print_exc(); return None

# --- 1.2 Time Series Visualization ---
def visualize_time_series(df, date_col, indicators, price_cols_prefix="", num_price_samples=5):
    """ Plots economic indicators and a sample of price columns over time. """
    print("\n--- 2. Time Series Visualization ---")
    if date_col not in df.columns or not pd.api.types.is_datetime64_any_dtype(df[date_col]):
        print(f"Error: Date column '{date_col}' not found or not datetime. Skipping plots.")
        return

    valid_indicators = [ind for ind in indicators if ind in df.columns]
    if valid_indicators:
        print("Plotting economic indicators..."); plt.figure(figsize=(15, 7))
        for indicator in valid_indicators: plt.plot(df[date_col], df[indicator], label=indicator)
        plt.title('Economic Indicators Over Time', fontsize=16); plt.xlabel('Date'); plt.ylabel('Value')
        plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()
    else: print("Warning: No valid economic indicators found.")

    price_columns = [col for col in df.columns if col.startswith(price_cols_prefix) and col not in indicators and col != date_col and pd.api.types.is_numeric_dtype(df[col])]
    if price_columns:
        print("Plotting sample housing prices..."); num_samples = min(len(price_columns), num_price_samples)
        if num_samples > 0:
            sample_regions = np.random.choice(price_columns, size=num_samples, replace=False)
            plt.figure(figsize=(15, 7))
            for region in sample_regions: plt.plot(df[date_col], df[region], label=f'Region {region}')
            plt.title(f'Sample Housing Prices Over Time ({num_samples} Random Regions)', fontsize=16); plt.xlabel('Date'); plt.ylabel('Housing Price')
            plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()
        else: print("No numeric price columns for sampling.")
    else: print("Warning: No numeric price columns found.")

# --- 1.3 Descriptive Statistics ---
def show_descriptive_stats(df, columns):
    """ Displays descriptive statistics for specified numerical columns. """
    print("\n--- 3. Descriptive Statistics ---")
    valid_columns = [col for col in columns if col in df.columns and pd.api.types.is_numeric_dtype(df[col])]
    if not valid_columns: print("Warning: No valid numeric columns for stats."); return
    try:
        print(f"Statistics for: {valid_columns}")
        print(df[valid_columns].describe().to_string(float_format="{:.2f}".format))
    except Exception as e: print(f"Error calculating stats: {e}")

# --- 1.4 Distribution Visualization (Histograms & Box Plots) ---
def visualize_distributions(df, columns):
    """ Generates histograms and box plots for specified columns. """
    print("\n--- 4. Distribution Visualization ---")
    valid_columns = [col for col in columns if col in df.columns and pd.api.types.is_numeric_dtype(df[col])]
    if not valid_columns: print("Warning: No valid numeric columns for plots."); return
    num_cols = len(valid_columns); n_rows = (num_cols + 1) // 2
    if num_cols == 0: return

    print("Generating Histograms..."); plt.figure(figsize=(14, 4 * n_rows))
    for i, col in enumerate(valid_columns):
        plt.subplot(n_rows, 2, i + 1)
        if df[col].isnull().all(): print(f"Skipping histogram for {col} (all NaNs)."); continue
        sns.histplot(df[col].dropna(), kde=True, bins=30, color=sns.color_palette("viridis", num_cols)[i])
        plt.title(f'Distribution of {col}'); plt.xlabel(col); plt.ylabel('Frequency')
    plt.tight_layout(pad=2.0); plt.show()

    print("Generating Box Plots..."); plt.figure(figsize=(14, 4 * n_rows))
    for i, col in enumerate(valid_columns):
        plt.subplot(n_rows, 2, i + 1)
        if df[col].isnull().all(): print(f"Skipping box plot for {col} (all NaNs)."); continue
        sns.boxplot(y=df[col].dropna(), color=sns.color_palette("magma", num_cols)[i])
        plt.title(f'Box Plot of {col}'); plt.ylabel(col)
    plt.tight_layout(pad=2.0); plt.show()

# --- 1.5 Correlation Analysis ---
def analyze_correlation(df, columns, title_suffix=""):
    """ Calculates and visualizes the correlation matrix for specified columns. """
    print(f"\n--- 5. Correlation Analysis {title_suffix} ---")
    valid_columns = [col for col in columns if col in df.columns and pd.api.types.is_numeric_dtype(df[col])]
    if len(valid_columns) < 2: print(f"Warning: Need >= 2 numeric columns for correlation. Found: {valid_columns}"); return
    try:
        correlation_matrix = df[valid_columns].corr()
        print("\nCorrelation Matrix:"); print(correlation_matrix.to_string(float_format="{:.2f}".format))
        plt.figure(figsize=(max(8, len(valid_columns)*0.8), max(6, len(valid_columns)*0.6)))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, annot_kws={"size": 8})
        plt.title(f'Correlation Matrix {title_suffix}', fontsize=14); plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0)
        plt.tight_layout(); plt.show()
    except Exception as e: print(f"Error during correlation analysis: {e}")

# --- 1.6 Pair Plot Visualization ---
def visualize_pairplot(df, columns):
    """ Generates pair plots to visualize pairwise relationships between features. """
    print("\n--- 6. Pair Plot Visualization ---")
    valid_columns = [col for col in columns if col in df.columns and pd.api.types.is_numeric_dtype(df[col])]
    if len(valid_columns) < 2: print(f"Warning: Need >= 2 numeric columns for pair plot. Found: {valid_columns}"); return
    try:
        print(f"Generating Pair Plot for: {valid_columns}")
        df_pairplot = df[valid_columns].dropna()
        if df_pairplot.empty: print("Warning: No data for pair plot after dropping NaNs."); return
        sample_df = df_pairplot.sample(min(len(df_pairplot), 1000)) if len(df_pairplot) > 1000 else df_pairplot
        pair_plot = sns.pairplot(sample_df, diag_kind='kde', plot_kws={'alpha': 0.6, 's': 20})
        pair_plot.fig.suptitle('Pair Plot of Economic Indicators (Sampled)', y=1.02, fontsize=16); plt.show()
    except Exception as e: print(f"Error during pair plot: {e}")

# --- 1.7 User Input for Target Region ---
def get_target_region_input(df, available_indicators, date_col_name, prefix=""):
    """ Prompts the user to enter a target region (zipcode) and validates it. """
    print("\n--- 7. Select Target Region ---")
    exclude_cols = set(available_indicators);
    if date_col_name in df.columns: exclude_cols.add(date_col_name)
    potential_regions = sorted([col for col in df.columns if col.startswith(prefix) and col not in exclude_cols and pd.api.types.is_numeric_dtype(df[col])])

    if not potential_regions:
        print("Warning: Could not auto-detect regions. Checking all numeric columns...");
        potential_regions = sorted([col for col in df.columns if col not in exclude_cols and pd.api.types.is_numeric_dtype(df[col])])
        if not potential_regions: print("Error: No suitable numeric target columns found."); return None
        print("Please choose from the following numeric columns (examples):"); print(potential_regions[:min(10, len(potential_regions))])

    while True:
        region_code = input(f"Enter target region code/name (e.g., 10001 or full name): ")
        target_column_with_prefix = f"{prefix}{region_code}"
        target_column_direct = region_code
        target_column = None
        if target_column_with_prefix in potential_regions: target_column = target_column_with_prefix
        elif target_column_direct in potential_regions: target_column = target_column_direct

        if target_column: print(f"Selected target region: {target_column}"); return target_column
        else:
            print(f"Error: '{region_code}' not found among valid regions.")
            print(f"Examples: {potential_regions[:min(5, len(potential_regions))]}")
            if input("Try again? (y/n): ").lower() != 'y': return None

# --- 1.8 Target Variable Analysis (Scatter Plots) ---
def visualize_target_relationships(df, features, target_col):
    """ Generates scatter plots comparing feature columns to the target variable. """
    print(f"\n--- 8. Feature vs. Target '{target_col}' Visualization ---")
    valid_features = [feat for feat in features if feat in df.columns and pd.api.types.is_numeric_dtype(df[feat])]
    if not valid_features: print("Warning: No valid numeric features for scatter plots."); return
    if target_col not in df.columns or not pd.api.types.is_numeric_dtype(df[target_col]): print(f"Warning: Target '{target_col}' not found/numeric."); return
    num_features = len(valid_features); n_rows = (num_features + 1) // 2
    if num_features == 0: return

    print(f"Generating scatter plots for features vs {target_col}...")
    plt.figure(figsize=(14, 5 * n_rows))
    for i, feature in enumerate(valid_features):
        plt.subplot(n_rows, 2, i + 1); plot_data = df[[feature, target_col]].dropna()
        if plot_data.empty: print(f"Skipping plot for {feature} (NaNs)."); continue
        sns.scatterplot(x=plot_data[feature], y=plot_data[target_col], alpha=0.5, s=20)
        sns.regplot(x=plot_data[feature], y=plot_data[target_col], scatter=False, color='red', line_kws={'linewidth':1})
        plt.title(f'{feature} vs {target_col}'); plt.xlabel(feature); plt.ylabel(f'Housing Price ({target_col})')
    plt.tight_layout(pad=2.0); plt.show()

# --- 1.9 Prepare Data for Model ---
def prepare_data_for_model(df, feature_cols, target_col):
    """ Prepares data for model, selecting features/target and handling NaNs. """
    print("\n--- 9. Preparing Data for Modeling ---")
    feature_cols_no_date = [col for col in feature_cols if col != DATE_COLUMN]
    valid_feature_cols = [col for col in feature_cols_no_date if col in df.columns and pd.api.types.is_numeric_dtype(df[col])]
    missing = set(feature_cols_no_date) - set(valid_feature_cols)
    if missing: print(f"Warning: Features not found/numeric: {missing}. Using: {valid_feature_cols}")
    if not valid_feature_cols: print("Error: No valid feature columns."); return None, None, None
    if target_col not in df.columns or not pd.api.types.is_numeric_dtype(df[target_col]): print(f"Error: Target '{target_col}' not found/numeric."); return None, None, None

    try:
        cols_to_check = valid_feature_cols + [target_col]; df_subset = df[cols_to_check]
        initial_rows = len(df_subset); df_model_ready = df_subset.dropna()
        rows_dropped = initial_rows - len(df_model_ready)
        if rows_dropped > 0: print(f"Dropped {rows_dropped} rows with NaNs in selected columns.")
        if df_model_ready.empty: print("Error: No data after dropping NaNs."); return None, None, None

        X = df_model_ready[valid_feature_cols].copy()
        # Scale target by 1000 for potentially better model performance/interpretation
        y = (df_model_ready[target_col] / 1000.0).values
        print("Features (X) selected:"); print(X.head())
        print("\nTarget (y) selected and scaled (/1000):"); print(y[:5])
        print("\nShapes: X:", X.shape, " y:", y.shape)
        return X, y, valid_feature_cols
    except Exception as e: print(f"Error preparing data: {e}"); traceback.print_exc(); return None, None, None

# --- 1.10 Data Splitting ---
def split_data(X, y, test_size=0.2, random_state=42):
    """ Splits the data into training and testing sets. """
    print("\n--- 10. Splitting Data ---")
    if X is None or y is None or len(X) != len(y) or len(X) == 0: print("Error: Invalid input for splitting."); return None, None, None, None
    if len(X) < 2: print(f"Error: Not enough data ({len(X)}) for split."); return None, None, None, None
    # Check for potential empty sets after split
    min_train_samples = 1
    min_test_samples = 1
    if int(test_size * len(X)) < min_test_samples or int((1-test_size)*len(X)) < min_train_samples :
         print(f"Warning: test_size={test_size} might lead to empty sets with {len(X)} samples.")
    try:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
        print(f"Data split: Train={len(X_train)}, Test={len(X_test)}")
        return X_train, X_test, y_train, y_test
    except ValueError as ve: print(f"Error splitting data (ValueError): {ve}"); return None, None, None, None
    except Exception as e: print(f"Unexpected error splitting data: {e}"); return None, None, None, None

# --- 1.11 Model Evaluation Function (Common post-fitting) ---
def evaluate_model_performance(model, model_name, X_test, y_test, X_train=None):
    """ Evaluates a trained model (sklearn or Keras), prints metrics, plots, returns results. """
    print(f"\n--- Evaluating: {model_name} ---")
    results = {}
    try:
        print("Making predictions on test set...")
        predictions = model.predict(X_test)
        y_test_flat = y_test.ravel()
        predictions_flat = predictions.ravel() # Keras predict might be (n, 1)

        print(f"\n--- {model_name}: Evaluation Metrics ---")
        mae = mean_absolute_error(y_test_flat, predictions_flat)
        mse = mean_squared_error(y_test_flat, predictions_flat)
        rmse = np.sqrt(mse); r2 = r2_score(y_test_flat, predictions_flat)
        adj_r2 = "N/A"
        # Calculate Adj R2 if possible (need num features 'p' from training data shape)
        if X_train is not None:
            n = X_test.shape[0]; p = X_train.shape[1]
            if n > p + 1: adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
            else: print(f"Warning: Cannot calculate Adj R (n={n} <= p+1={p+1}).")
        else: print("Warning: X_train shape not provided, cannot calculate Adj R.")

        print(f"MAE:    {mae:.4f}"); print(f"MSE:    {mse:.4f}"); print(f"RMSE:   {rmse:.4f}")
        print(f"R:     {r2:.4f}"); print(f"Adj R: {adj_r2 if isinstance(adj_r2, str) else f'{adj_r2:.4f}'}")
        results.update({'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2': r2, 'Adj_R2': adj_r2, 'Predictions': predictions_flat})

        # --- Diagnostic Plots ---
        print("\nGenerating Diagnostic Plots..."); plt.figure(figsize=(14, 6))
        plt.subplot(1, 2, 1); plt.scatter(y_test_flat, predictions_flat, alpha=0.6, s=25, label='Predictions')
        min_val = min(y_test_flat.min(), predictions_flat.min()); max_val = max(y_test_flat.max(), predictions_flat.max())
        plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', linewidth=1, label='Ideal Fit')
        plt.xlabel('Actual Values (Scaled/1k)'); plt.ylabel('Predicted Values (Scaled/1k)'); plt.title(f'{model_name}: Actual vs. Predicted'); plt.legend(); plt.grid(True)
        plt.subplot(1, 2, 2); residuals = y_test_flat - predictions_flat
        plt.scatter(predictions_flat, residuals, alpha=0.6, s=25); plt.axhline(y=0, color='red', linestyle='--', linewidth=1)
        plt.xlabel('Predicted Values (Scaled/1k)'); plt.ylabel('Residuals'); plt.title(f'{model_name}: Residual Plot'); plt.grid(True)
        plt.tight_layout(pad=2.0); plt.show()

        print(f"\n--- {model_name}: Evaluation Complete ---"); return results
    except Exception as e: print(f"Error during {model_name} evaluation: {e}"); traceback.print_exc(); return None

# --- 1.12 Neural Network Model Definition (Accepts HParams) ---
def build_nn_model(input_shape, learning_rate=0.001, units1=64, units2=32):
    """Builds a Keras MLP model, adaptable for manual grid search."""
    tf.keras.backend.clear_session() # Clear previous models from memory
    model = keras.Sequential([
        layers.Input(shape=(input_shape,), name="Input_Layer"),
        layers.Dense(units1, activation='relu', name='Dense_1'),
        layers.Dense(units2, activation='relu', name='Dense_2'),
        layers.Dense(1, name='Output_Layer')
    ])
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['mae', 'mse'])
    return model

# --- 1.13 Manual NN Grid Search Function (NEW) ---
def manual_nn_grid_search(X_train_scaled, y_train, param_grid, cv_folds):
    """ Performs manual grid search with K-Fold CV for a Keras model. """
    print(f"\n--- Running Manual NN Grid Search (CV={cv_folds}) ---")
    kfold = KFold(n_splits=cv_folds, shuffle=True, random_state=42)
    results = []
    best_score = -np.inf # Using R2, higher is better
    best_params = None
    input_shape = X_train_scaled.shape[1]

    # Generate parameter combinations
    keys, values = zip(*param_grid.items())
    param_combinations = [dict(zip(keys, v)) for v in itertools.product(*values)]
    print(f"Testing {len(param_combinations)} parameter combinations...")
    total_fits = len(param_combinations) * cv_folds
    fit_count = 0

    start_time_grid = time.time()

    for i, params in enumerate(param_combinations):
        fold_scores = []
        print(f"\nCombination {i+1}/{len(param_combinations)}: {params}")
        start_time_comb = time.time()

        for fold, (train_idx, val_idx) in enumerate(kfold.split(X_train_scaled)):
            fit_count += 1
            print(f"  Fold {fold+1}/{cv_folds} (Fit {fit_count}/{total_fits})... ", end="")
            X_fold_train, X_fold_val = X_train_scaled[train_idx], X_train_scaled[val_idx]
            y_fold_train, y_fold_val = y_train[train_idx], y_train[val_idx]

            try:
                # Build and compile model for this fold and params
                model = build_nn_model(input_shape=input_shape, **params)

                # Use EarlyStopping based on validation loss during fold training
                early_stopping = tf.keras.callbacks.EarlyStopping(
                    monitor='val_loss', patience=10, restore_best_weights=True, verbose=0
                )

                # Train the model
                history = model.fit(
                    X_fold_train, y_fold_train,
                    epochs=NN_MANUAL_EPOCHS, # Use fixed epochs for grid search phase
                    batch_size=NN_MANUAL_BATCH_SIZE, # Use fixed batch size
                    validation_data=(X_fold_val, y_fold_val), # Use fold validation set
                    callbacks=[early_stopping],
                    verbose=0 # Suppress epoch output for cleaner grid search log
                )

                # Evaluate on validation set using R2 score
                y_pred_val = model.predict(X_fold_val).ravel()
                score = r2_score(y_fold_val.ravel(), y_pred_val)
                fold_scores.append(score)
                print(f"Val R: {score:.4f}")

            except Exception as e:
                print(f"Error in fold {fold+1} for params {params}: {e}")
                # Assign a very poor score if a fold fails
                fold_scores.append(-np.inf)


        avg_score = np.mean(fold_scores) if fold_scores else -np.inf
        print(f"  Avg R across folds: {avg_score:.4f} (Time: {time.time() - start_time_comb:.2f}s)")
        results.append({'params': params, 'avg_score': avg_score})

        # Update best score and params
        if avg_score > best_score:
            best_score = avg_score
            best_params = params
            print(f"  *** New Best Score Found! ***")

    print(f"\n--- Manual NN Grid Search Complete (Total time: {time.time() - start_time_grid:.2f}s) ---")
    if best_params:
        print(f"Best parameters found: {best_params}")
        print(f"Best average cross-validation R score: {best_score:.4f}")
    else:
        print("Manual grid search did not find any successful parameter combinations.")

    return best_params


# --- 1.14 Model Comparison Function ---
# (Keep the function compare_model_performance as it was for 3 models)
def compare_model_performance(lr_results, rf_results, nn_results, y_test_actual):
    """ Compares the performance of LR, RF, and NN models. """
    print("\n--- 12. Model Performance Comparison ---")
    models_present = {}
    if lr_results: models_present['Linear Regression'] = lr_results
    if rf_results: models_present['Random Forest'] = rf_results
    if nn_results: models_present['Neural Network'] = nn_results

    if not models_present:
        print("No model results available for comparison.")
        return

    # --- Metrics Comparison Table ---
    metrics_data = {'Metric': ['MAE', 'MSE', 'RMSE', 'R2', 'Adj_R2']}
    for name, results in models_present.items():
         metrics_data[name] = [
            results.get('MAE'), results.get('MSE'), results.get('RMSE'),
            results.get('R2'), results.get('Adj_R2')
        ]

    metrics_df = pd.DataFrame(metrics_data)

    # Format the numeric columns
    float_format_func = lambda x: f"{x:.4f}" if isinstance(x, (int, float)) else str(x)
    for col in models_present.keys():
        metrics_df[col] = metrics_df[col].apply(float_format_func)

    print("\nEvaluation Metrics Summary:")
    print(metrics_df.to_string(index=False))

    # --- Comparison Plots ---
    print("\nGenerating Comparison Plots...")
    y_test_flat = y_test_actual.ravel()

    plt.figure(figsize=(15, 7))

    # 1. Actual vs. Predicted
    plt.subplot(1, 2, 1)
    colors = {'Linear Regression': 'blue', 'Random Forest': 'green', 'Neural Network': 'orange'}
    preds = {}
    all_preds = []
    all_actual = [y_test_flat.min(), y_test_flat.max()]

    for name, results in models_present.items():
        pred = results['Predictions']
        preds[name] = pred
        all_preds.extend([pred.min(), pred.max()])
        plt.scatter(y_test_flat, pred, alpha=0.5, s=20, color=colors[name], label=f'{name[:2]} Preds')

    min_val = min(all_actual + all_preds)
    max_val = max(all_actual + all_preds)
    plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', linewidth=1, label='Ideal Fit')
    plt.xlabel('Actual Values (Scaled/1k)'); plt.ylabel('Predicted Values (Scaled/1k)'); plt.title('Comparison: Actual vs. Predicted'); plt.legend(); plt.grid(True)

    # 2. Residuals vs. Predicted
    plt.subplot(1, 2, 2)
    for name, results in models_present.items():
        pred = results['Predictions']
        residuals = y_test_flat - pred
        plt.scatter(pred, residuals, alpha=0.5, s=20, color=colors[name], label=f'{name[:2]} Residuals')

    plt.axhline(y=0, color='red', linestyle='--', linewidth=1)
    plt.xlabel('Predicted Values (Scaled/1k)'); plt.ylabel('Residuals'); plt.title('Comparison: Residual Plot'); plt.legend(); plt.grid(True)

    plt.tight_layout(pad=2.0); plt.show()
    print("\n--- Model Comparison Complete ---")


# --- Main Execution Flow ---
if __name__ == "__main__":
    # 1. Load Data
    df_processed = load_inspect_convert_dates(FILE_PATH, DATE_COLUMN)

    if df_processed is not None:
        # 2-8. EDA & Target Selection
        visualize_time_series(df_processed, DATE_COLUMN, ECONOMIC_INDICATORS, TARGET_PREFIX)
        show_descriptive_stats(df_processed, ECONOMIC_INDICATORS)
        visualize_distributions(df_processed, ECONOMIC_INDICATORS)
        analyze_correlation(df_processed, ECONOMIC_INDICATORS, title_suffix="of Economic Indicators")
        visualize_pairplot(df_processed, ECONOMIC_INDICATORS)
        target_region_column = get_target_region_input(df_processed, ECONOMIC_INDICATORS, DATE_COLUMN, TARGET_PREFIX)

        if target_region_column:
            analyze_correlation(df_processed, ECONOMIC_INDICATORS + [target_region_column], title_suffix=f"including Target '{target_region_column}'")
            visualize_target_relationships(df_processed, ECONOMIC_INDICATORS, target_region_column)

            # 9. Prepare Data
            X, y, actual_feature_names = prepare_data_for_model(df_processed, MODEL_FEATURES, target_region_column)

            if X is not None and y is not None:
                # 10. Split Data
                X_train, X_test, y_train, y_test = split_data(X, y)

                if X_train is not None and X_test is not None:
                    # 10b. Scale Features
                    print("\n--- 10b. Scaling Features ---")
                    scaler = StandardScaler()
                    X_train_scaled = scaler.fit_transform(X_train)
                    X_test_scaled = scaler.transform(X_test)
                    print("Features scaled using StandardScaler.")
                    print(f"Scaled Train Shape: {X_train_scaled.shape}, Scaled Test Shape: {X_test_scaled.shape}")

                    # Initialize results
                    lr_results, rf_results, nn_results = None, None, None
                    best_rf_model, final_nn_model = None, None
                    best_nn_params = None

                    # --- 11a. Linear Regression (No Tuning) ---
                    print("\n--- 11a. Training & Evaluating Linear Regression ---")
                    try:
                        lr_model = LinearRegression()
                        lr_model.fit(X_train_scaled, y_train.ravel())
                        lr_results = evaluate_model_performance(lr_model, "Linear Regression", X_test_scaled, y_test, X_train_scaled)
                    except Exception as e: print(f"Linear Regression failed: {e}"); traceback.print_exc()

                    # --- 11b. Random Forest Tuning (GridSearchCV) ---
                    print("\n--- 11b. Tuning Random Forest with GridSearchCV ---")
                    rf_param_grid = {
                        'n_estimators': [50, 100],
                        'max_depth': [10, 20, None],
                        'min_samples_split': [5, 10],
                        'min_samples_leaf': [3, 5],
                        'max_features': ['sqrt', 1.0]
                    }
                    try:
                        base_rf = RandomForestRegressor(random_state=42)
                        grid_search_rf = GridSearchCV(estimator=base_rf, param_grid=rf_param_grid, cv=CV_FOLDS, scoring='r2', n_jobs=-1, verbose=1) # Using R2 score
                        start_time = time.time()
                        grid_search_rf.fit(X_train_scaled, y_train.ravel())
                        print(f"RF Grid Search took: {time.time() - start_time:.2f} seconds")
                        print(f"Best RF Score (R): {grid_search_rf.best_score_:.4f}")
                        print(f"Best RF Parameters: {grid_search_rf.best_params_}")
                        best_rf_model = grid_search_rf.best_estimator_
                        # Evaluate the best RF model
                        rf_results = evaluate_model_performance(best_rf_model, "Random Forest (Tuned)", X_test_scaled, y_test, X_train_scaled)
                    except Exception as e: print(f"Random Forest GridSearchCV failed: {e}"); traceback.print_exc()

                    # --- 11c. Neural Network Tuning (Manual Grid Search) ---
                    print("\n--- 11c. Tuning Neural Network with Manual Grid Search ---")
                    # Define smaller grid for manual search example
                    nn_param_grid_manual = {
                         'learning_rate': [0.01, 0.001],
                         'units1': [32, 64],
                         'units2': [16, 32]
                    }
                    try:
                         best_nn_params = manual_nn_grid_search(X_train_scaled, y_train.ravel(), nn_param_grid_manual, cv_folds=CV_FOLDS)

                         # Train final NN model with best params found
                         if best_nn_params:
                             print("\nTraining final Neural Network model with best parameters...")
                             final_nn_model = build_nn_model(input_shape=X_train_scaled.shape[1], **best_nn_params)
                             es_final = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, verbose=1)

                             history_final = final_nn_model.fit(
                                 X_train_scaled, y_train.ravel(),
                                 epochs=FINAL_NN_EPOCHS, # Train for potentially more epochs
                                 batch_size=NN_MANUAL_BATCH_SIZE, # Can use fixed or tune this too
                                 validation_split=0.2, # Use validation split for early stopping
                                 callbacks=[es_final],
                                 verbose=0 # Can set to 1 to see final training progress
                             )
                             print("Final NN model training complete.")
                             # Evaluate the final NN model
                             nn_results = evaluate_model_performance(final_nn_model, "Neural Network (Tuned)", X_test_scaled, y_test, X_train_scaled)
                         else:
                             print("Skipping final NN training and evaluation as no best parameters were found.")

                    except Exception as e: print(f"Neural Network Manual Search/Final Training failed: {e}"); traceback.print_exc()


                    # --- 12. Compare Models ---
                    compare_model_performance(lr_results, rf_results, nn_results, y_test)

                else:
                    print("\nSkipping modeling due to error in data splitting.")
            else:
                print("\nSkipping modeling due to error in data preparation.")
        else:
            print("\nNo target region selected. Skipping target-specific analysis and modeling.")
    else:
         print("\nScript halted due to error during data loading.")

    print("\n--- Data Exploration, Model Tuning, and Comparison Script Finished ---")
```
